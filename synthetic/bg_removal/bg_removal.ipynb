```python
from withoutbg import WithoutBG
import os
from pathlib import Path

# Initialize model once - loaded into memory
model = WithoutBG.opensource()

def show_progress(progress):
    print(f"Processing: {progress * 100:.0f}%")

image_dir = "/home/lukasb/Documents/NoisyLabelDefectDetection/data/gan_data/corrosion"
output_dir = Path("/home/lukasb/Documents/NoisyLabelDefectDetection/data/stylegan/corrosion_withoutbg/")
output_dir.mkdir(parents=True, exist_ok=True)

valid_exts = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp"}
images = []

for name in sorted(os.listdir(image_dir)):
    path = os.path.join(image_dir, name)
    if os.path.isfile(path) and os.path.splitext(name)[1].lower() in valid_exts:
        images.append(path)

# Don't let withoutbg save to JPEG (JPEG can't store alpha). Save ourselves as PNG instead.
results = model.remove_background_batch(images)
for in_path, image in zip(images, results):
    out_path = output_dir / f"{Path(in_path).stem}.png"
    image.save(out_path)
# Returns list of PIL Image objects in `results`
```
