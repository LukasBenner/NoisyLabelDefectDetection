```python
import os
import shutil
from pathlib import Path

# Source directory
source_dir = Path("../data/new_data/2026_01_09")
# Destination directory
dest_dir = Path("../data/unlabeled_ssl_data")

# Create destination directory if it doesn't exist
dest_dir.mkdir(parents=True, exist_ok=True)

# Find all png and jpg files recursively
image_extensions = ['*.png', '*.jpg', '*.jpeg', '*.PNG', '*.JPG', '*.JPEG']
copied_count = 0

for ext in image_extensions:
    for image_file in source_dir.rglob(ext):
        # Copy file to destination
        shutil.copy2(image_file, dest_dir / image_file.name)
        copied_count += 1

print(f"Copied {copied_count} image files to {dest_dir}")
```
