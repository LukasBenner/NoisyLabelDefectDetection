```python
import sys
sys.path.append('..')  # Add parent directory to path

from torchvision.datasets import ImageFolder
import numpy as np
```

```python
ds = ImageFolder("../data/surface/all")
```

```python
# Initialize accumulators for sum and sum of squares
r_sum, g_sum, b_sum = 0, 0, 0
r_sq_sum, g_sq_sum, b_sq_sum = 0, 0, 0
total_pixels = 0

for img, _ in ds:
    # Convert PIL image to numpy array and normalize to [0, 1]
    img_array = np.array(img) / 255.0
    
    # Accumulate sums
    r_sum += img_array[:, :, 0].sum()
    g_sum += img_array[:, :, 1].sum()
    b_sum += img_array[:, :, 2].sum()
    
    # Accumulate squared sums
    r_sq_sum += (img_array[:, :, 0] ** 2).sum()
    g_sq_sum += (img_array[:, :, 1] ** 2).sum()
    b_sq_sum += (img_array[:, :, 2] ** 2).sum()
    
    total_pixels += img_array.shape[0] * img_array.shape[1]

# Calculate mean
r_mean = r_sum / total_pixels
g_mean = g_sum / total_pixels
b_mean = b_sum / total_pixels

# Calculate std
r_std = np.sqrt(r_sq_sum / total_pixels - r_mean ** 2)
g_std = np.sqrt(g_sq_sum / total_pixels - g_mean ** 2)
b_std = np.sqrt(b_sq_sum / total_pixels - b_mean ** 2)

print(f"Mean: R={r_mean:.4f}, G={g_mean:.4f}, B={b_mean:.4f}")
print(f"Std:  R={r_std:.4f}, G={g_std:.4f}, B={b_std:.4f}")
print(f"\nFor PyTorch normalization:")
print(f"mean=[{r_mean:.4f}, {g_mean:.4f}, {b_mean:.4f}]")
print(f"std=[{r_std:.4f}, {g_std:.4f}, {b_std:.4f}]")
```
