# Split data in training set and hold back test set

The training data is split in train and test set by removing 15% of the traing data stratified by the class labels.
The test set is moved to a separate folder.

```python
import os
import shutil
import numpy as np
from sklearn.model_selection import train_test_split
from torchvision import datasets
```

```python
source_dir = "../data/surface/train"
test_dir = "../data/surface/test_uncleaned"

# Create test directory if it doesn't exist
os.makedirs(test_dir, exist_ok=True)
```

```python
dataset = datasets.ImageFolder(source_dir)
all_indices = list(range(len(dataset)))
all_targets = [dataset.samples[i][1] for i in all_indices]
all_paths = [dataset.samples[i][0] for i in all_indices]
class_names = dataset.classes



for i in range(4):
    train_indices, test_indices = train_test_split(
        all_indices,
        test_size=0.05,
        random_state=i+np.random.randint(10,30),
        stratify=all_targets)
    
    for idx in test_indices:
        src_path = all_paths[idx]
        target = all_targets[idx]
        class_name = class_names[target]
        new_path = os.path.join(test_dir, class_name)
        os.makedirs(new_path, exist_ok=True)
        file_name = os.path.basename(src_path)
        dst_path = os.path.join(new_path, file_name)
        shutil.copy(src_path, dst_path)
```

Now clean the test data manually by removing images that do not belong the the respective class.

```python
test_dir = "../data/surface/test_new"
source_dir = "../data/surface/train_new"
dataset = datasets.ImageFolder(source_dir)
all_indices = list(range(len(dataset)))
all_targets = [dataset.samples[i][1] for i in all_indices]
all_paths = [dataset.samples[i][0] for i in all_indices]
class_names = dataset.classes
```

```python
test_images = set()
for class_name in class_names:
    test_class_dir = os.path.join(test_dir, class_name)
    if os.path.exists(test_class_dir):
        for img_name in os.listdir(test_class_dir):
            test_images.add(img_name)

# Remove test images from training set
for class_name in class_names:
    train_class_dir = os.path.join(source_dir, class_name)
    if os.path.exists(train_class_dir):
        for img_name in os.listdir(train_class_dir):
            if img_name in test_images:
                img_path = os.path.join(train_class_dir, img_name)
                os.remove(img_path)
                print(f"Removed {img_path}")
```
